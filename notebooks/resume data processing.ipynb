```python
bpo_df = pd.DataFrame({
    'Category': ['BPO'] * len(bpo_samples),  
    'Feature': bpo_samples
})

automobile_df = pd.DataFrame({
    'Category': ['AUTOMOBILE'] * len(automobile_samples),  
    'Feature': automobile_samples
})

data = pd.concat([data, bpo_df, automobile_df], ignore_index=False)
data.reset_index(drop=True, inplace=True)
```

```python
from sklearn.utils import resample
import pandas as pd

balanced_data = []
target_count = 100

for category in data['Category'].unique():
    category_data = data[data['Category'] == category]
    
    if len(category_data) < target_count:
        balanced_category_data = resample(category_data, replace=True, n_samples=target_count, random_state=42)
    else:
        balanced_category_data = category_data
    balanced_data.append(balanced_category_data)

balanced_data = pd.concat(balanced_data, ignore_index=True)
```

```python
print(balanced_data['Category'].value_counts())
```

```python
balanced_data.isnull().sum()
```

```python
import re

def cleanResume(txt):
    cleanText = re.sub('http\S+\s', ' ', txt)  # Remove URLs
    cleanText = re.sub('#\S+\s', ' ', cleanText)  # Remove hashtags
    cleanText = re.sub('@\S+', '  ', cleanText)  # Remove mentions
    cleanText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleanText)  # Remove punctuation
    cleanText = re.sub(r'[^\x00-\x7f]', ' ', cleanText)  # Remove non-ASCII characters
    cleanText = re.sub('\s+', ' ', cleanText)  # Remove extra spaces
    return cleanText
```

```python
balanced_data['Feature'] = balanced_data['Feature'].apply(cleanResume)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

X = balanced_data['Feature']  
y = balanced_data['Category']  

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y) 

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
import pickle

with open("data/X_train_tfidf.pkl", "wb") as f:
    pickle.dump(X_train_tfidf, f)
with open("data/X_test_tfidf.pkl", "wb") as f:
    pickle.dump(X_test_tfidf, f)
with open("data/y_train.pkl", "wb") as f:
    pickle.dump(y_train, f)
with open("data/y_test.pkl", "wb") as f:
    pickle.dump(y_test, f)
with open('data/class_names.txt', 'w') as f:
    for class_name in label_encoder.classes_:
        f.write(f"{class_name}\n")
with open("models/tfidf_vectorizer_categorization.pkl", "wb") as f:
    pickle.dump(tfidf_vectorizer, f)
```
